```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
df = pd.read_csv("Titanic-Dataset.csv")
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['PassengerId'], df['Fare'])
plt.xlabel("Passenger ID")
plt.ylabel("Fare")
plt.title("Line Plot - Fare over Passenger ID")
plt.show()
```

```python
# scatter plot - age vs fare
plt.figure(figsize=(6,4))
plt.scatter(df['Age'], df['Fare'])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter Plot - Age vs Fare")
plt.show()
```

```python
#histogram - age distribution
plt.figure(figsize=(6,4))
plt.hist(df['Age'], bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram - Age Distribution")
plt.show()
```

```python
#bar plot - survival count
survival = df['Survived'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(survival.index, survival.values)
plt.xticks([0,1], ['Died', 'Survived'])
plt.xlabel("Survival Status")
plt.ylabel("Count")
plt.title("Bar Chart - Survival Count")
plt.show()
```

```python
#pie chart - gender distribution
gender_counts = df['Sex'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title("Pie Chart - Gender Distribution")
plt.show()
```

```python
#box plot - age by passenger class
plt.figure(figsize=(6,4))
df.boxplot(column='Age', by='Pclass')
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.title("Age Distribution by Class")
plt.suptitle("")
plt.show()

```

```python
#area plot - cumulative fare
plt.figure(figsize=(10,4))
df_sorted = df.sort_values("PassengerId")
plt.fill_between(df_sorted['PassengerId'], df_sorted['Fare'], alpha=0.5)
plt.xlabel("Passenger ID")
plt.ylabel("Fare")
plt.title("Area Plot - Cumulative Fare Trend")
plt.show()
```

```python
#heatmap - correlation matrix
plt.figure(figsize=(8,6))
corr = df[['Age','Fare','Survived','Pclass','SibSp','Parch']].corr()
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Heatmap - Correlation Matrix")
plt.show()
```

```python
#subplots - age and fare distribution
fig, axs = plt.subplots(1, 2, figsize=(12,5))

axs[0].hist(df['Age'], bins=20)
axs[0].set_title("Age Distribution")

axs[1].hist(df['Fare'], bins=20)
axs[1].set_title("Fare Distribution")

plt.show()
```

```python
#stacked bar chart - survival by gender
survival_gender = df.groupby(['Sex','Survived']).size().unstack()

plt.figure(figsize=(6,4))
plt.bar(survival_gender.index, survival_gender[0], label="Died")
plt.bar(survival_gender.index, survival_gender[1], bottom=survival_gender[0], label="Survived")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Stacked Bar Chart - Survival by Gender")
plt.legend()
plt.show()

```

SEABORN 

```python
import seaborn as sns
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"].dropna(), bins=30, kde=True)
plt.title("Age Distribution of Titanic Passengers")
plt.show()
```

```python
#count plot - survival by class
plt.figure(figsize=(8,5))
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.show()
```

```python
#Bar Plot (with grouping)
plt.figure(figsize=(8,5))
sns.barplot(x="Pclass", y="Fare", data=df)
plt.title("Average Fare by Passenger Class")
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df)
plt.title("Age vs Fare (Colored by Survival)")
plt.show()
```

```python
#Box Plot
plt.figure(figsize=(8,5))
sns.boxplot(x="Pclass", y="Age", data=df)
plt.title("Age Distribution by Class")
plt.show()
```

```python
#Violin Plot
plt.figure(figsize=(8,5))
sns.violinplot(x="Pclass", y="Age", data=df)
plt.title("Age Distribution by Class (Violin Plot)")
plt.show()
```

```python
#heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#pairplot
sns.pairplot(df[["Age", "Fare", "Pclass", "Survived"]].dropna(), hue="Survived")
plt.show()
```

```python
#kde plot 
plt.figure(figsize=(8,5))
sns.kdeplot(df["Fare"].dropna(), shade=True)
plt.title("Fare Density Plot")
plt.show()
```
